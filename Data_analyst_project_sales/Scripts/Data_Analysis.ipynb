```python
import pandas as pd
# 1. Load the data 
path = 'C:\\Users\\srish\\OneDrive\\Desktop\\Data_analyst_project_sales\\Raw_data\\sales_data_sample.csv'
df = pd.read_csv(path, encoding='latin1')
# 2. Basic Cleaning
print("Missing values per column:")
print(df.isnull().sum())
# Remove duplicates
df.drop_duplicates(inplace=True)
# 3. Data Transformation 
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
df['Profit_Margin'] = (df['SALES'] - (df['QUANTITYORDERED'] * df['PRICEEACH'])) / df['SALES']
# 4. Save the cleaned file
output_path = 'C:\\Users\\srish\\OneDrive\\Desktop\\Data_analyst_project_sales\\Processed_data\\cleaned_sales_data.csv'
df.to_csv(output_path, index=False)
print("\n--- SUCCESS ---")
print("Cleaned data saved to Processed_data folder!")
```
